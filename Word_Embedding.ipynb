```python
import os
```

```python
l = []
for root, dirs, files in os.walk("data", topdown=False):
    for name in files:
        print(name)
        with open(os.path.join(root, name), mode='r', errors='ignore', encoding = 'utf-8') as f:
            l.append(f.read())
```

```python
import pandas as pd
df = pd.DataFrame(l, columns=['texte'])
```

### Suppression du bruit et tokenisation

```python
# tout en minuscule
df['texte'] = df['texte'].map(lambda x: x.lower())
#df['texte']
```

```python
# suppression des accents
def remove_accents(a):
    return unidecode.unidecode(a)
```

```python
df['texte'] = df['texte'].apply(remove_accents)
```

```python
df['texte']
```

```python
# Tokenization des phrases en mot
from nltk import RegexpTokenizer
toknizer = RegexpTokenizer(r'''\w'|\w+|[^\w\s]''')
token = df.apply(lambda row: toknizer.tokenize(row['texte']),axis= 1)
token
```

```python
# Suppression de la ponctuation
import string
import re 
#string.punctuation
token_1 = token.apply(lambda x: [item.strip(string.punctuation) for item in x])
token_2 = token_1.apply(lambda x: [re.sub(r'\’', '', item) for item in x])
token_2
```

```python
# suppression des nombres
token_4 = token_2.apply(lambda x: [re.sub(r'\d', '', item) for item in x])
token_4 
```

```python
# Stopword Removal : Suppression des mots vides
from nltk.corpus import stopwords 
import nltk 
nltk.download('stopwords')
stop_words = set(stopwords.words('french')) 
stop_token = token_4.apply(lambda x: [item for item in x if item not in stop_words])

stop_token
```

```python
# On rajoute à notre dataframe les données noramlisées comme nouvelle colonne
df['texte_norm'] = stop_token
df.head()
```

## Formation sur le modèle Genism word2vec

```python
# pour le word2vec
from gensim.test.utils import common_texts, get_tmpfile
from gensim.models import Word2Vec
model = Word2Vec(df['texte_norm'], min_count=1,size= 50,workers=3, window =3, sg = 1)
```
