### 3) Nettoyage spécifique pour de l’analyse de texte

```python
pip install unidecode
```

```python
# mettre en minuscule
# enlever les accents
# enlever les chiffres et caracteres spéciaux

# enlever les stop words


import unidecode
import re

def nettoyage(texte):
    tex=[]
    # Construction de la liste de stop words
    import stop_words
    sw_1=stop_words.get_stop_words('fr')
    from nltk.corpus import stopwords
    sw_nltk = set(stopwords.words('french')) 
    sw=list(set(sw_1+list(sw_nltk)))+['vais','vous','je','les','qu','ils','ou']
    
    texte=unidecode.unidecode(texte.lower())

    
    p="([a-z]{1,})"
    for elem in re.findall(p,texte):
        if elem in sw or elem==' ':
            continue
        else:
            tex.append(elem)
    return ' '.join(tex)

df['commentaire'][10],nettoyage(df['commentaire'][10])
```

```python
df['commentaire']=df['commentaire'].apply(nettoyage)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['commentaire'])
print(vectorizer.get_feature_names())

print(X.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
#vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['commentaire'])
print(vectorizer.get_feature_names())
```

```python
#Definition d'un seuil à partir duquel les elements (tf idf) sont pertinents
seuil=X.max()/4

tfidf=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names())
final_tfidf=[]
for i in tfidf.index:
    tmp=tfidf.loc[i]
    t=[]
    while(tmp.max() > seuil):
        #print(i,tmp[tmp.idxmax(axis=1)],tmp.idxmax(axis=1))
        t.append(tmp.idxmax(axis=1))
        del tmp[tmp.idxmax(axis=1)]
    final_tfidf.append(' '.join(t))
    continue
```
